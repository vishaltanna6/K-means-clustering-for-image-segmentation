# Importing matplotlib library

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

# Reading an image of cars on road

```python
car = imread('car.jpg')
```

# Displaying image

```python
plt.figure(figsize=(14,14))
plt.imshow(car)
```

# Scaling the data

```python
car = car/255.0
```

```python
car.shape
```

# Reshaping the data 

```python
X = car.reshape(-1,3)
```

```python
X.shape
```

# Image Segmentation 

```python
from sklearn.cluster import KMeans
```

# Mention the number of cluster to be formed and fit the data to the mdoel

```python
km = KMeans(n_clusters=3)
km.fit(X)
```

# Found 3 cluster with the RGB values as below 

```python
km.cluster_centers_
```

```python
img_seg = km.cluster_centers_
```

```python
img_seg = img_seg[km.labels_]
```

```python
img_seg = img_seg.reshape(car.shape)
```

```python
plt.figure(figsize=(14,14))
plt.imshow(img_seg)
```
